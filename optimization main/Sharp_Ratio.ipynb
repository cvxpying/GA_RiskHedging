```python
import numpy as np
import pandas as pd
import random
```

배당금 데이터 가져오기

```python
dividends = pd.read_csv('/content/dividend.csv')
print(dividends)
```

```python
# 종목 이름을 원소로 갖는 List 생성
stocks = []
for i in range(len(dividends.columns)-1):
    stocks.append(dividends.columns[i+1])
print(stocks)
```

현재 없는 데이터 Random하게 생성

```python
# 주식 가격 데이터 (월별)
np.random.seed(0)

prices_dict = {}

for i in range(64):
    series_name = stocks[i]
    prices_dict[series_name] = pd.Series(np.random.randn(122).cumsum() + 100000)

# DataFrame으로 변환
prices = pd.DataFrame(prices_dict)
#prices.insert(0, 'Date', dividends['Date']) # df 사이즈 맞춰주기
print(prices)
```

R_P(포트폴리오 기대수익률) 정의

```python
# A = 기업별 배당금 수익률 (월별 배당금 / 월별로 처리된 수정종가)
# B = 기업별 총 수익률 (주식 가격 수익률 + 배당금 수익률)
# C = 총 수익률 = A + B
# D = 기업별 평균 수익률 ( sum(C) / 시점 수)
# R_P = sum(기업별 평균 수익률 * 각 종목별 비중 W_i)

# 최종적으로 기업별 평균 수익률을 구하면 끝임!!!!!
# 주가 수익률 계산

mean_return_dict = {}

for i in range(64):
    series_name = stocks[i]
    returns = pd.Series(prices[stocks[i]]).pct_change().dropna()
    dividend_yields = pd.Series(dividends[stocks[i]] / prices[stocks[i]].shift(1))


    result = np.mean(returns + dividend_yields[1:])
    mean_return_dict[series_name] = result


# DataFrame으로 변환
mean_return = pd.DataFrame.from_dict(mean_return_dict, orient='index', columns=['Value']).T
print(mean_return)

# mean_return 에 각 기업별 가중치 W_i를 곱하면 R_P 값이 나오게 된다 !!!

# 예시)
random_data = np.random.rand(64)
normalized_data = random_data / random_data.sum()
print(mean_return*normalized_data)

R_P = np.mean()
```

R_F 정의

```python
R_F = 0.01
```

시그마_P 정의

```python
merged_df = pd.concat([prices.drop(columns = "Date"), dividends.drop(columns = "Date")], axis = 0)
all_profit = merged_df.groupby(level=0).sum()


g_i = all_profit.pct_change()
print(g_i)
print(g_i.cov())
```

```python
annual_g_i = g_i.cov()*122
# 참고한 자료에서 공분산 행렬에 우리가 분석할 시점의 개수만큼 곱하였는데 그 이유에 대해선 모르겠음.
```

샤프함수 정의

```python
def sharp(*args):
    weights = []
    for i in args:
        weights.append(i)
    weights = np.array(weights)
    R_P = np.mean(mean_return*weights)
    risk = np.sqrt(np.dot(weights.T, np.dot(annual_g_i, weights)))
    sharp = ((R_P - R_F)/risk)
    return sharp

sharp_ratio = []

for i in range(200):
    random_data = np.random.rand(64)
    random_normalized = random_data / random_data.sum()
    sharp_ratio.append(sharp(*random_normalized))

print([x for x in sharp_ratio if x > 0])
## 샤프지수가 음수로밖에 안 나오는 문제가 생김
## R_P가 아주 작은 값으로 나오고 있는데 R_P에서 문제가 있는 것 같음.
```
