## R_p

 = sum[(배당금/환산종가) + ((현시점환산종가 - 이전시점환산종가)/이전시점환산종가)] / 시점수 * 각 기업 가중치

= sum(avg_returns * 각 기업 가중치)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
avg_returns = pd.read_excel('avg_returns.xlsx')

avg_returns
```

```python
df = avg_returns.copy()

df['Date'] = pd.to_datetime(df['Date'])

df['Date'] = df['Date'].dt.year

df
```

```python

def calculate_Rp(df, seed=42):
    np.random.seed(seed)
    tickers = df.columns[1:]  # 'Date'열 제외
    # num_tickers = len(tickers)
    num_tickers = 2

    selected_tickers = np.random.choice(tickers, num_tickers, replace=False)
    weights = np.random.random(num_tickers)
    weights /= weights.sum()  # 가중치 합 = 1로 조정
    
    annual_returns = df[selected_tickers]
    weighted_returns = annual_returns.multiply(weights, axis=1)
    Rp = weighted_returns.sum(axis=1)
    
    return Rp

Rp = calculate_Rp(df)

for year, rp in zip(df['Date'], Rp):
    print(f"Year: {year}, R_p: {rp:.6f}")
```
